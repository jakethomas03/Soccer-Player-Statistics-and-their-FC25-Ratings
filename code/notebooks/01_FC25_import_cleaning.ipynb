**Description:** Downloading the FC25 Dataset into a pandas dataframe and removes unnecessary columns.

```python
# import libraries
import pandas as pd
```

```python
fc25_df = pd.read_csv("/Users/Jake/Documents/GitHub/Data-Wrangling-Project/data/raw/FC25 Dataset/male_players.csv", encoding="utf-8")
display(fc25_df)
```

```python
# Remove unnecessary columns
new_df = fc25_df.loc[:, ["Name", "OVR", "PAC", "SHO", "PAS", "DRI", "DEF", "PHY", "Nation", "League", "Team", "Age", "Height", "Weight", "Position"]]
display(new_df)
```

```python
# Replace notable players that are spelled differently between datasets

# Change vini jr to Vinicius Júnior
new_df["Name"] = new_df["Name"].str.replace("Vini Jr.", "Vinicius Júnior")

# Change Gabriel to Gabriel Magalhães
new_df["Name"] = new_df["Name"].str.replace("Gabriel", "Gabriel Magalhães")

# Change Pierre-Emile Højbjerg to Pierre Højbjerg
new_df["Name"] = new_df["Name"].str.replace("Pierre-Emile Højbjerg", "Pierre Højbjerg")

# Change Heung Min Son to Son Heung-min
new_df["Name"] = new_df["Name"].str.replace("Heung Min Son", "Son Heung-min")

# Change Fran García to Fran Garcia
new_df["Name"] = new_df["Name"].str.replace("Fran García", "Fran Garcia")

# Change Julio César Enciso to Julio Enciso
new_df["Name"] = new_df["Name"].str.replace("Julio César Enciso", "Julio Enciso")

# Change Brahim to Brahim Díaz
new_df["Name"] = new_df["Name"].str.replace("Brahim", "SBrahim Díaz")

# Change Carvajal to Dani Carvajal
new_df["Name"] = new_df["Name"].str.replace("Carvajal", "Dani Carvajal")

# Change Balde to Alejandro Balde
new_df["Name"] = new_df["Name"].str.replace("Balde", "Alejandro Balde")

# Change Ronald Araujo to Ronald Araújo
new_df["Name"] = new_df["Name"].str.replace("Ronald Araujo", "Ronald Araújo")

# Change Romeo Lavia to Roméo Lavia
new_df["Name"] = new_df["Name"].str.replace("Romeo Lavia", "Roméo Lavia")

# Change Oyarzabal to Mikel Oyarzabal
new_df["Name"] = new_df["Name"].str.replace("Oyarzabal", "Mikel Oyarzabal")

# Change Palhinha to João Palhinha
new_df["Name"] = new_df["Name"].str.replace("Palhinha", "João Palhinha")

# Change Savinho to Sávio
new_df["Name"] = new_df["Name"].str.replace("Savinho", "Sávio")

# Change Benjamin White to Ben White
new_df["Name"] = new_df["Name"].str.replace("Benjamin White", "Ben White")

# Change Jérémy Doku to Jeremy Doku
new_df["Name"] = new_df["Name"].str.replace("Jérémy Doku", "Jeremy Doku")

# Change Zubimendi to Martín Zubimendi
new_df["Name"] = new_df["Name"].str.replace("Zubimendi", "Martín Zubimendi")

# Change Giovanni Reyna to Gio Reyna
new_df["Name"] = new_df["Name"].str.replace("Giovanni Reyna", "Gio Reyna")

# Change Ferran Torres to Ferrán Torres
new_df["Name"] = new_df["Name"].str.replace("Ferran Torres", "Ferrán Torres")
```

```python
# Save cleaned dataframe as new csv
new_df.to_csv("fc25_data_clean.csv", index=False, encoding="utf-8")
```
